```python
# 이 코드는 네이버 지식에서 1번째 페이지 부터 n번째 페이지 까지 게시글 링크를 받아온다.

import requests, time, re
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

mainUrl = 'https://kin.naver.com/search/list.naver?query='
pageUrl = '&page='
keyword = input('지식인 검색어 입력') # 검색어 입력
maxPage = 10

linkUrlLst = []

for pageNum in tqdm(range(1, maxPage + 1)):
    url = mainUrl + keyword + pageUrl + str(pageNum)
    html = requests.get(url)
    if html.status_code == 200:
        soup = BeautifulSoup(html.text, 'html.parser')

        ulTag = soup.find('ul', class_='basic1') # 게시글 목록
        titleLST = ulTag.find_all('dl')
        for linkUrl in titleLST:
            linkUrlLst.append(linkUrl.find('dt').find('a')['href'])

        time.sleep(0.1)
    else:
        print(f'응답을 받지 못했습니다. 응답메시지{html.status_code}')
```

```python
# 게시글 링크에 접속하여 게시글 내용 가져오기

titLst = [] # 제목
cntLst = [] # 본문
anwLst = [] # 답변

for suburl in tqdm(linkUrlLst):
    time.sleep(0.1)
    subHtml = requests.get(suburl) # 게시글 링크에 요청 보내기
    if subHtml.status_code == 200:
        # 게시글 가져오기
        subsoup = BeautifulSoup(subHtml.text, 'html.parser')

        # 제목
        try:
            title = subsoup.find('div', class_='title').text
            title = re.sub('[^ㄱ-ㅣ가-힣a-zA-Z ] ', '', title)

        except:
            print(suburl)
            continue
        
        # 본문
        try:
            content = subsoup.find('div', class_='c-heading__content').text
            content = re.sub('[^ㄱ-ㅣ가-힣a-zA-Z ]', '', content)

        except:
            content = title

        # 답변
        try:
            answer = ''
            answerLst = subsoup.find_all('div', class_='se-module se-module-text')
            for txt in answerLst:
                answer = answer + ' ' + txt.text
            answer = re.sub('[^ㄱ-ㅣ가-힣a-zA-Z ] ', '', answer)
                
        except:
            answer = ''

        # 리스트에 추가
        titLst.append(title)
        cntLst.append(content)
        anwLst.append(answer)

    else:
        print(f'응답을 받지 못했습니다. 응답메시지{html.status_code}')

df = pd.DataFrame({'title': titLst, 'content': cntLst, 'answer': anwLst})
df.to_csv(keyword + '지식인크롤링.csv', encoding='utf-8')
```
