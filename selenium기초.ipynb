```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time, re

driver = webdriver.Chrome('chromedriver.exe')
keyword = input('블로그 검색어 입력: ')

url1 = 'https://section.blog.naver.com/Search/Post.naver?pageNo='
url2 = '&rangeType=ALL&orderBy=sim&keyword='
maxPage = 10
blogUrlList = []

for pageNum in range(1, maxPage + 1):
    url = url1 + str(pageNum) + url2 + keyword
    driver.get(url)
    time.sleep(0.5)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    linkLst = soup.find_all('a', class_='desc_inner')
    for url in linkLst:
        blogUrlList.append(url['href'])

driver.close()
print(len(blogUrlList))
print(blogUrlList)
```

```python

titLst = [] # 제목
cntLst = [] # 본문

title = ''
content = ''
driver = webdriver.Chrome('chromedriver.exe')

for blog in blogUrlList:
    try:
        driver.get(blog)
        time.sleep(0.5)
        driver.switch_to_frame('mainFrame')
        subSoup = BeautifulSoup(driver.page_source, 'html.parser')

        # 제목
        title = subSoup.find('span', class_='se-fs-').text
        title = re.sub('[^ㄱ-ㅣ가-힣a-zA-Z ] ', '', title)

        # 본문
        content = subSoup.find('div', class_='se-main-container').text
        content = re.sub('[^ㄱ-ㅣ가-힣a-zA-Z ] ', '', content)
    except:
        pass

    titLst.append(title)
    cntLst.append(content)
driver.close()
```
